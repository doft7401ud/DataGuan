```python
import os
import numpy as np
import pandas as pd
import torch
```

```python
# data = torch.empty()

ds_root = 'D:\data_guan_cut_sub\DS'
smarteye_root = 'D:\data_guan_cut_sub\interpolated'
output_root = r'D:\data_guan_cut_sub\final'

# ds_files = []
ds_files = ['No38', 'No39', 'No40', 'No41', 'No42', 'No43']
# for file in files:
#     ds_files.append(os.path.join(ds_root, file))

# ds_files = os.listdir(ds_root)
# smarteye_files = os.listdir(smarteye_root+'\\No1')

# for file in ds_files:
#     ds_file = os.path.join(ds_root, file, f'{file}_DS_eachLaneChange.xlsx')
#     ds_data = pd.read_excel(ds_file,header=None,sheet_name=None)
#     for ds_sheet_name, ds_df in ds_data.items():
#         for smarteye_file in smarteye_files:
#             smarteye_file_path = os.path.join(smarteye_root, file, smarteye_file)
#             smarteye_file_data = pd.read_excel(smarteye_file_path, sheet_name = None, header = None)
#             for smarteye_sheet_name, smarteye_df in smarteye_file_data.items():
#                 print(smarteye_df.shape[0] == ds_df.shape[0])
for file in ds_files:
    ds_file = os.path.join(ds_root, file, f'{file}_DS_eachLaneChange.xlsx')
    ds_data = pd.read_excel(ds_file, header=None, sheet_name=None)
    smarteye_files = os.listdir(os.path.join(smarteye_root, file))
    for ds_sheet_name, ds_df in ds_data.items():
        for smarteye_file in smarteye_files:
            smarteye_file_path = os.path.join(smarteye_root, file, smarteye_file)
            smarteye_data = pd.read_excel(smarteye_file_path, header=None, sheet_name=None)
            smarteye_file = list(smarteye_file.split('_'))[1:-2]
            col_name = ''.join(smarteye_file)
            if ds_sheet_name in smarteye_data:
                smarteye_df = smarteye_data[ds_sheet_name]
                smarteye_df.columns = [col_name]
                ds_df = pd.concat([ds_df,smarteye_df], axis=1)
        ds_df.to_csv(os.path.join(output_root,f'{file[2:]}_{ds_sheet_name[5:]}.csv'))
        print(f'File {file}_{ds_sheet_name} Finished')
```
