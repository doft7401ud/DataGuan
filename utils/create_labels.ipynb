```python
import os
import pandas as pd
import json

base_folder_path = 'D:\data_guan_cut_sub\Questionnaires'
csv_folder_path = r'D:\data_guan_cut_sub\final'

final_labels = {}

for participant_id in range(1, 44): 
    participant_folder = os.path.join(base_folder_path, f"No{participant_id}")
    
    excel_file_path = os.path.join(participant_folder, f"No{participant_id}_Subjective Questionnaires.xlsx")
    
    if os.path.exists(excel_file_path):
        df = pd.read_excel(excel_file_path, header=None)
        # 假设第二列是 STS-AD 分数，即第 1 索引列
        sts_ad_scores = df[1]  # 获取第 2 列 (index = 1)
        
        # 计算中位数
        median_sts_ad = sts_ad_scores.median()
        
        labels = []
        
        for experiment_id, row in df.iterrows():
            sts_ad_score = row[1]  # 第二列是 STS-AD 分数
            
            # 二分类根据 STS-AD 中位数
            if sts_ad_score <= median_sts_ad:
                label = False  # 小于等于中位数
            else:
                label = True  # 大于中位数
            
            # 存入列表
            labels.append({
                "experiment": f"{participant_id}_{experiment_id + 1}.csv",
                "label": label
            })
        
        final_labels[f"Participant_{participant_id}"] = labels
    else:
        print(f"Excel 文件 {excel_file_path} 不存在，跳过此文件。")

# 将结果保存为json文件
json_output_path = '.\\label_2.json'  # 修改后的输出文件名
with open(json_output_path, 'w') as json_file:
    json.dump(final_labels, json_file, indent=4)

print(f"分类标签已保存为 {json_output_path}")
```
